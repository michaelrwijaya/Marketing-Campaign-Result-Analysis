### Market Basket using Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
product_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Mengubah data menjadi format biner
binary_data = data[product_columns].applymap(lambda x: 1 if x > 0 else 0)
binary_data.head()
```

```python
# Menerapkan algoritma Apriori
frequent_itemsets = apriori(binary_data, min_support=0.05, use_colnames=True)

# Menghitung aturan asosiasi
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules = rules.sort_values(by='lift', ascending=False)
rules
```

Setelah itu, kami mengkhususkan pencarian Antecedents dan Consequents hanya  untuk per satu tipe produk saja, tidak digabung agar lebih membuat paket bundling lebih spesifik.

```python
# Filter aturan asosiasi untuk satu kolom
target_column = 'MntFruits'
filtered_rules = rules[rules['consequents'] == frozenset([target_column])]

# Menampilkan hasil
filtered_rules
```

Dari hasil association_rules diatas, kita dapat melihat bahwa produk MntFishProducts dan MntSweetProducts adalah produk yang memiliki keterkaitan tertinggi dimana kedua produk ini memiliki nilai **lift** tertinggi. Semakin tinggi nilai lift, semakin tinggi juga asosiasi antar produk. Jika nilai lift lebih dari 1, hal tersebut dapat kita katakan bahwa kedua item tersebut saling terkait.

Selain itu, dapat dilihat bahwa nilai support dari produk MntFishProducts dan MntSweetProducts  adalah sebesar 0.724107 yang berarti terdapat 0.724107% dari total transaksi bahwa 2 produk tersebut dijual bersamaan.

Sedangkan dari kolom antecedent dan consequent support menjelaskan bahwa jika antecedent lebih tinggi dari consequent, maka aturan yang akan diterapkan adalah aturan nomor 1 (bukan nomor 2), dan sebaliknya. Dalam hal ini, nilai anteseden lebih tinggi dari nilai konsekuen. Ini berarti kami akan menerapkan aturan nomor 1, yaitu **MntFishProducts â†’ MntSweetProducts**. ### Customer Segmentation using Clustering Algorithm: K-Means

Untuk metode clustering ini, kami melakukan sedikit penyesuaian agar dapat memberikan segmentasi pelanggan dengan baik, salah satunya dengan dimentional reduction

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
data2 = data.copy()
```

```python
categorical = data2.select_dtypes(include=["object"])

le = LabelEncoder()
for col in categorical:
  data2[col] = le.fit_transform(data2[col])

data2.dtypes
```

Melabeli kolom agar proses clustering lebih optimal dan mudah

```python
sc = StandardScaler()

data2 = pd.DataFrame(sc.fit_transform(data2), columns=data2.columns)
data2
```

```python
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(data2)
principalDf = pd.DataFrame(data = principalComponents, columns=['PC1', 'PC2', 'PC3'])

principalDf
```

```python
def visualize_clustering(data, predictions=None, marker_size=3, title="Principal Components in 3-Dimensions"):
  if predictions is not None:
    data['preds'] = predictions
    fig = px.scatter_3d(data, x='PC1', y='PC2', z='PC3', color='preds',
                      color_discrete_sequence=px.colors.qualitative.Pastel, opacity=0.5)
  else:
    fig = px.scatter_3d(data, x='PC1', y='PC2', z='PC3',
                        color_discrete_sequence=px.colors.qualitative.Pastel, opacity=0.5)

  fig.update_traces(marker_size = marker_size)
  fig.update_layout(title_text=title, title_x=0.5)
  fig.show()
```

Melakukan dimensionality reduction, karena dataset ini memiliki banyak kolom, dan beberapa kolom memiliki korelasi antar kolom yang tinggi. Hal ini akan membuat clustering menjadi lebih mudah.

```python
visualizer = KElbowVisualizer(KMeans(), k=15);

visualizer.fit(principalDf);
visualizer.show();
```

Berdasarkan visualisasi elbow method di atas, jumlah optimal dari cluster adalah 5.

```python
cluster_k = KMeans(n_clusters=5)
cluster_k.fit(principalDf)

visualize_clustering(principalDf, cluster_k.labels_, title="Clustering KMeans | K=5")
```

Di atas adalah hasil clustering menggunakan K-Means dengan jumlah K=5

Selanjutnya, kami melakukan analisis untuk menjawab platform mana yang paling populer di kalangan lima cluster tersebut

```python
data['Cluster'] = cluster_k.labels_

data.head()
```

```python
accepted = data.groupby("Cluster")[['NumWebPurchases', 'NumStorePurchases', 'NumCatalogPurchases']].mean().T

accepted.plot(kind='bar', rot=0);
set_labels(x="Platform", y="Rata-Rata", title="Rata-Rata Pembelian berdasarkan Platform per Cluster");
```

Berdasarkan visualisasi di atas, dapat diketahui bahwa cluster 3 merupakan cluster tertinggi yang membeli melalui platform web, diikuti cluster 4, 1, 2, kemudian 0.

Selanjutnya, kami menganalisis cluster mana yang menerima paling banyak promosi secara rata-rata

```python
accepted = data.groupby("Cluster")[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4',
           'AcceptedCmp5', 'Response']].mean()

accepted.plot(kind='bar', rot=0);
set_labels(x="Cluster", y="Rata-Rata Jumlah Penawaran Diterima", title="Rata-Rata Jumlah Penawaran Diterima per Cluster");
```

Berdasarkan visualisasi di atas, dapat diketahui bahwa cluster 4 menerima penawaran promosi terbanyak, jauh dari yang lainnya. Kemudian diikuti cluster 3.

Selanjutnya, kami juga mencari tahu kategori produk mana yang paling populer di kalangan cluster

```python
cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
accepted = data.groupby("Cluster")[cols].mean().T

accepted.plot(kind='bar', rot=0);
set_labels(x="Kategori", y="Jumlah", title="Rata-Rata Jumlah Pengeluaran Berdasarkan Kategori per Cluster");
```

Dapat kita ketahui bahwa walaupun cluster 4 memiliki jumlah pelanggan yang sedikit, mereka seringkali membeli wines. Untuk cluster 1, mereka merupakan pembeli produk buah, ikan, permen, dan emas terbanyak.
